Todas las edades estén en formato númerico (las que sean 0 o NaN, sustituirlas por el valor medio de la columna EDAD).

```python
#df = pd.read_csv("personas_sucio.csv")
```

```python
df['edad'] = pd.to_numeric(df['edad'], errors='coerce') #Convertir a numerico
df
```

```python
#Calculamos la media
edad_media = df[df['edad'].notna() & (df['edad'] > 0)]['edad'].mean()
edad_media = round(edad_media)
edad_media
df
```

```python
#Asignamos la media
df['edad'] = df['edad'].fillna(edad_media)
df.loc[df['edad'] == 0, 'edad'] = edad_media
df
```

 Todas las ciudades estén escritas en mayúsculas.

```python
df['ciudad'] = df['ciudad'].str.strip().str.upper()
df
```

Todos los emails tengan el formato correcto (texto@otrotexto.com)

```python
#df = pd.read_csv("personas_sucio.csv")
```

```python
df = df[df["email"].str.contains(r"^[\w\.-]+@[\w\.-]+\.\w+$", na=False)]
df
```

Todas las fechas estén en formato DD/MM/AAAA.

```python
df["fecha"] = pd.to_datetime(df.get("fecha"), errors="coerce")
df
```
