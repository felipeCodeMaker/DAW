5. Cambiamos nombres de columna, los espacios pasan de ' ' a '_' snake_case

```python
#Cambiamos nombres de columna, los espacios pasan de ' ' a '_'

def snake_case(column_name):
    #Convierte los nombres de las columnas a snake_case, les pone una _ donde habia ' '.
    return column_name.lower().replace(' ', '_').replace('-', '_')

# Aplicar la conversi√≥n a todas las columnas
df.columns = [snake_case(col) for col in df.columns]

print("\n Nombres de columnas en snake_case:")
print(df.columns.tolist())
```

6. Detectar y tratar valores nulos

```python
#Detectamos el numero de nulos que hay por columna
null_counts = df.isna().sum()
print('contador de nulos \n', null_counts)
print("Valores nulos por columna:")
print(null_counts[null_counts > 0])  # Solo mostramos columnas con valores nulos
```

```python
#Cambiamos todos los valores nulos por otros, seg√∫n el tipo de dato

# Para columnas num√©ricas, rellenar con 0 o mediana
numeric_columns = df.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    if df[col].isna().sum() > 0:
        # Usamos mediana para evitar efecto de outliers
        df[col] = df[col].fillna(0)     #Se supone que si pongo: df[col].median() rellena en vex de con 0 con la mediana
        #print(f"Columna num√©rica '{col}': valores nulos rellenados con mediana")

# Para columnas categ√≥ricas, rellenar con 'Unknown'
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna('Unknown')
        #print(f"Columna categ√≥rica '{col}': valores nulos rellenados con 'Unknown'")
```

7. Eliminar duplicados

```python
df = df.drop_duplicates()
```

8. Limpiar textos y normalizar valores

```python
# Columnas a normalizar (ajusta seg√∫n tu dataset)
text_columns_to_normalize = ['neighbourhood', 'room_type', 'city', 'NAME', 'host_name']
#.astype(str) --> Convierte todos los valores de la columna a tipo string√ß
#.str.strip() --> Elimina espacios en blanco al principio y al final de cada string
#.str.lower() --> Convierte todo el texto a min√∫sculas

for col in text_columns_to_normalize:
    if col in df.columns:
        # Convertir a min√∫sculas y eliminar espacios en blanco
        df[col] = df[col].astype(str).str.strip().str.lower()
        #print(f"Columna '{col}' normalizada")
        
print(df)
```

9. Limpiar precios

```python
print(df['price'])
```

```python
df["price"] = df["price"].replace('Unknown', '0', regex=True)
df["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)
print(df['price'])
```
